```
 
grade=35
if grade >= 90:
    print("A+ grade")
 
elif grade >=80:
    print("A grade")
 
elif grade >=70:
    print("B+ grade")
 
elif grade >= 60:
    print("B grade")
 
 elif grade >= 50:
    print("C grade")
 
elif grade >= 35:
    print("D grade")
else:
    print("Failing grade")
```

```
 
 
lower = 1
upper = 1000
 
print("Prime numbers between", lower, "and", upper, "are:")
 
for num in range(lower, upper + 1):
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num)
```
